## DataFrame Merge

```python
# 판다스 패키지 로딩
import pandas as pd
```

### Inner Join

```python
# [데이터 merge 실습] 두개의 데이터 프레임 생성

data1 = {
    '학번' : [1, 2, 3, 4],
    '이름' : ['아이유', '김연아', '홍길동', '강감찬'],
    '학과' : ['철학', '경역학', '컴퓨터', '물리학']
}

data2 = {
    '학번' : [1,2,4,5],
    '학년' : [2,4,3,1],
    '학점' : [1.5,2.0,4.1,3.8]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

display(df1)
display(df2)
```

```python
# inner join 
inner_df = pd.merge(df1,df2, on='학번', how='inner')
display(inner_df)
```

```python
### Outer Join
outer_df = pd.merge(df1,df2, on = '학번', how='outer')
display(outer_df)
```

```python
def func_b(a, key):
    i=0
    while True:
        if i ==len(a):
            return -1
        if a[i] == key:
            return i
        i+=1
a = [2,5,1,3,9,6,7]
index = func_b(a,3)
if index == -1:
    print('None')
else:
    print(f'a[{index}]')
```

### Left and Right Join

```python
left_df = pd.merge(df1,df2, on='학번', how='left')
display(left_df)
```

```python
right_df = pd.merge(df1,df2, on='학번', how='right')
display(right_df)
```

### merge응용: 현실 데이터를 다룰때 생길만 한 경우

1) 같은 의미의 다른 컬럼명을 병합할때

```python
#case1: 같은 의미 다른 컬럼명  ---> 학번 vs 학생고유번호
data1 = {
    '학번' : [1, 2, 3, 4],
    '이름' : ['아이유', '김연아', '홍길동', '강감찬'],
    '학과' : ['철학', '경역학', '컴퓨터', '물리학']
}

data2 = {
    '학생고유번호' : [1,2,4,5],
    '학년' : [2,4,3,1],
    '학점' : [1.5,2.0,4.1,3.8]
}
df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

display(df1)
display(df2)
```

```python
merged_df = pd.merge(df1, df2, left_on='학번', right_on='학생고유번호', how='inner')
display(merged_df)
```

```python
# 두개의 공통 키 중 하나를 삭제
merged_df.drop('학생고유번호',axis=1, inplace=True)
display(merged_df)
```
